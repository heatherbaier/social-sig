```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
import matplotlib.image as mpimg
import pandas as pd
```

```python
for i in range(0, 120):
    df = pd.read_csv("./figs/im" + str(i) + ".csv")
    df["0"] = df["0"].str.split("(").str[1].str.split(",").str[0].astype(float)
    # plt.imshow(np.reshape(np.array(df["0"]), (224,224)))
    mpimg.imsave("./pics/im" + str(i) + ".png", np.reshape(np.array(df["0"]), (224,224)))
```
